# Downloading Data

```python
from fastbook import *
def search_images_ddg_corrected(term, max_images=200):
    "Search for `term` with DuckDuckGo and return a unique urls of about `max_images` images"
    assert max_images<1000
    url = 'https://duckduckgo.com/'
    res = urlread(url,data={'q':term})
    searchObj = re.search(r'vqd=([\d-]+)\&', res)
    assert searchObj
    requestUrl = url + 'i.js'
    params = dict(l='us-en', o='json', q=term, vqd=searchObj.group(1), f=',,,', p='1', v7exp='a')
    urls,data = set(),{'next':1}
    while len(urls)<max_images and 'next' in data:
        try:
            data = urljson(requestUrl,data=params)
            urls.update(L(data['results']).itemgot('image'))
            requestUrl = url + data['next']
        except (URLError,HTTPError): pass
        time.sleep(1)
    return L(urls)
#urls = search_images_ddg_corrected("cubist art", max_images=100)
#len(urls),urls[:30]
```

# Data Block

```python
path = 'gdrive/MyDrive/Art/'
```

```python
from fastbook import *      #only if haven't imported while downloading data
from fastai.vision import *
from fastai.metrics import error_rate
```

```python
dls = ImageDataLoaders.from_folder(path, valid_pct=0.2, bs=32, item_tfms=Resize(224))
dls.valid_ds.items[:3]
```

```python
dls.show_batch(nrows=1, figsize=(7,6))
```

# Model

```python
learn = cnn_learner(dls, models.resnet34, metrics=error_rate)
```

```python
learn.model
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

# Checking results & unfreeze transing on Resnet34

```python
interp_here = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp_here.top_losses()

len(dls.valid_ds)==len(losses)==len(idxs)
```

```python
interp_here.plot_top_losses(9, figsize=(15,11))
```

```python
interp_here.plot_confusion_matrix(figsize=(12,12), dpi=60)
```
